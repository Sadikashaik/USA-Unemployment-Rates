```python
# importing libraries

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
# plotting the bar graph for unemployment rate over the years

plt.figure(figsize = (8,23))
yearly_montly_sales = sns.catplot(x='Year', y='all', data=data,col='Month',kind='bar',palette='muted', col_wrap = 2,height=5,aspect=3)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Year',fontweight='bold',fontsize=23)
plt.ylabel('Unemployement Rate',fontweight='bold',fontsize=23)
plt.show()
```

* The unemployement rate is more in the year 2010, 2011 and 2020, 2021 till the month of August

```python
# plotting the bar graph for unemployment rate based on gender

sns.catplot(x='Year', y='women', data=data, palette='husl', kind='bar', aspect = 3)
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Unemployment of Women over the Years')
plt.show()

sns.catplot(x='Year', y='men', data=data, palette='dark', kind='bar', aspect = 3)
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Unemployment of Men over the Years')
plt.show()
```

* From the year 2009 to 2012 the unemployment rate of men is more compared to women  
* In the early 2000's and From the year 2013 unemployement rate for both men and women are almost equal

```python
# plotting the bar graph for unemployment rate based on Race

sns.relplot(x='Year', y='black', data=data, kind='line', color='red', aspect = 3)
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Unemployment of Black_People over the Years')
plt.show()

sns.relplot(x='Year', y='white', data=data, kind='line', color='green', aspect = 3)
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Unemployment of Whie_People over the Years')
plt.show()

sns.relplot(x='Year', y='hispanic', data=data, kind='line', aspect = 3)
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Unemployment of Hispanic over the Years')
plt.show()
```

* The unemployement rate for Black people is more and for white people is less all over the years.
* The unemployement rate for Hispanic people is more compare to white people

```python
# plotting the bar graph for unemployment rate based on Education

sns.relplot(x='Year', y='less_than_hs', data=data, kind = 'scatter', color = 'red')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Qualification below HS')
plt.show()

sns.relplot(x='Year', y='high_school', data=data, kind = 'scatter', color='orange')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Qualification is HS ')
plt.show()

sns.relplot(x='Year', y='some_college', data=data, kind = 'scatter', color='yellow')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('Qualified from Some College')
plt.show()

sns.relplot(x='Year', y="bachelor's_degree", data=data, kind = 'scatter', color='blue')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title("People with Bachelor's Degree")
plt.show()

sns.relplot(x='Year', y='advanced_degree', data=data, kind = 'scatter',color='green')
plt.xlabel('Year')
plt.ylabel('Unemployement Rate')
plt.title('People with Advanced Degree')
plt.show()

```

* The unemployement rate for people who completed their High School and people with only High School qualification is more  
* The unemployement rate for people with Advanced and Bachelor's degree is less

```python
# Creating a new dataframe with Year and Age columns

data_1 = data[['Year','16-24','25-54','55-64']]
```

```python
data_1
```

```python
# unemployment rate over the years based on age

yearly_data = data_1.groupby('Year').mean()
yearly_data.reset_index(inplace=True)
```

```python
print(yearly_data)
```

```python
# unemployment rate over the years based on age in a tabular format

from tabulate import tabulate

headers = ['Years','16-24','25-54','55-64']
table = tabulate(yearly_data,headers,tablefmt="fancy_grid")
```

```python
print(table)
```
